## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

**Decision:** Which content pages should be reviewed and refreshed first — out of thousands — given that an editorial team only has time to act on a handful each week.

**Who acts and what they do:** An editor or content strategist. They open the ranked queue, look at the top pages, and decide which ones to rewrite, expand, update, or monitor. Without the ranking, they'd either guess or review pages in no particular order.

**Cost of a wrong call:**
*   **False positive (flagging a healthy page):** Wasted editor hours on a page that didn't need attention.
*   **False negative (missing a truly declining page):** A page keeps losing traffic and rankings while no one acts — a slower but more costly mistake, since recovery gets harder the longer it's ignored.



## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

Opening the starter CSV reveals a few numbers that make this lane worth pursuing. Out of 30,000 rows across 32 clients, 1,205 rows have avg_position = 0 — meaning roughly 4% of the dataset has no search position data at all, not a rank of zero. A model that treats that column naively would be wrong on those rows immediately, so careful handling of missingness is already a real concern, not a theoretical one.

The rate columns also carry a quiet trap: ctr = 0.76 means 0.76%, not 76%. With CTR being one of the core signals for this lane, getting that wrong would flip the entire scoring logic upside down.

Finally, the pre-run starter pipeline shows that a random forest model correctly identified about 37 of the top 50 pages to review, compared to only 12 for a fixed baseline rule. That jump — from 12 to 37 right out of 50 — is the clearest evidence that learning from the signals in this dataset beats guessing with a hand-written rule, and it's the core promise this lane is built on.

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

What this work can honestly say is that pages ranked higher by the model were more likely to be flagged as currently declining in the starter dataset, and that combining signals like position, CTR, staleness, and engagement outperforms a single fixed rule — measured by how many of the top 50 recommendations actually turned out to be real candidates. These are observed, directional findings that support a human decision, not a guarantee. What this work cannot claim is that refreshing a flagged page caused any recovery, because proving causation would require a controlled experiment that this data alone cannot provide. It also cannot claim to reveal how Google's algorithm works, nor can it treat trend_direction or trend_pct as meaningful features — both are derived from the label itself, making them leakage risks that get excluded entirely. Finally, the label used here reflects the current 90-day window, not a future outcome, so any claim about "predicting" decline should be read as decision-support for human review, not as a forecast.
